# Les Catherines

Les Catherines are a trio blending the warm sounds of acoustic instruments with a rich mix of languages (English, French, Spanish, and more). Their eclectic style, inspired by world music and soulful interpretations, creates a captivating, multicultural listening experience.

![Les Catherines live](images/les-catherines-live.jpg)


## Our Music

From intimate acoustic arrangements to vibrant multilingual interpretations, our music blends styles, cultures, and voices. Each performance is an invitation to travel across languages, emotions, and musical traditions.

## Live Performances

As a trio, we perform regularly across Boulder County and the Denver area, bringing acoustic music to both intimate venues and public spaces.

[See upcoming events →](events)

## Music Community

In addition to our performances, we host the CBArt Studios Open Mic in Superior. CBArt Studios Open Mic is a warm and welcoming space where local musicians and audiences come together in a shared musical experience.
