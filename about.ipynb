
# ABOUT US

---

![Les Catherines live](images/les-catherines-root.jpg)

Les Catherines is an acoustic trio united by a love of reinterpretation. Rather than focusing on original compositions, we explore and reshape existing songs—bringing them into new spaces through voice, language, and arrangement.

Our repertoire includes adaptations of well-known songs such as “Feeling Good,” “Wish You Were Here,” and “In My Life,” alongside lesser-known pieces that we enjoy bringing into the light. We often revisit songs across languages—English, French, Spanish, and more—letting each version take on its own color and emotion.

The result is a wide and varied repertoire, where each song feels distinct and carries its own atmosphere. Our goal is for every performance to offer something different—familiar yet renewed, and never quite the same.

Our approach is simple: strip songs down to their essence, then rebuild them with warmth, harmony, and a sense of intimacy. Whether familiar or newly discovered, each piece becomes part of a shared musical moment.
