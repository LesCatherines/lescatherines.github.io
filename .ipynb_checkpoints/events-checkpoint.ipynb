# EVENTS

Les Catherines perform across Boulder County and the Denver area, bringing acoustic music to both intimate venues and cultural spaces. Alongside these performances, we also foster a local music community through the CBArt Studios Open Mic in Superior.

----

## Les Catherines Performances

We perform regularly in Boulder County, including CBArt Studios in Superior and Tandoori Grill in Boulder, and have also been invited to perform at venues such as the Alliance Française de Denver and the Cherry Creek Shopping Center. Our performances feature a multilingual acoustic repertoire, blending voices, styles, and interpretations into a warm and engaging musical experience.

### Alliance Française de Denver
April 25, 2026 · Denver, Colorado

An evening of acoustic music across languages and traditions.

Additional performances will be announced soon.

## Music Community — CBArt Studios Open Mic

In addition to performing, we host the CBArt Studios Open Mic in Superior, a recurring gathering where musicians and listeners come together in a relaxed and supportive setting.

An evening dedicated to sharing music, discovering new voices, and creating a sense of connection within the local community.
