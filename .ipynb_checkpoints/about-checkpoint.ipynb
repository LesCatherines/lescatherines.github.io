
# ABOUT US

---

![Les Catherines live](images/les-catherines-root.jpg)



Les Catherines is an acoustic trio born from a simple question: "what would this song sound like if we started over from scratch?"

We don’t just play covers; we listen, explore arrangements until something feels right for our voices and instruments. Sometimes an arrangement comes together quickly; other times it takes a while. But when it finally lands, it tends to stay... until we change our minds again.

Our repertoire includes well-known songs such as *Feeling Good*, *Wish You Were Here*, and *In My Life*, alongside lesser-known pieces that we enjoy bringing into the light. We let each song find the language, color, and atmosphere that suit it best.

The result is a varied repertoire in which each song carries its own atmosphere. We play for the moment when a listener recognizes a song and hears it differently at the same time. That is what we are chasing: a shared moment of emotion, connection, and surprise.
