# Les Catherines

Les Catherines are a trio blending the warm sounds of acoustic instruments with a rich mix of languages—English, French, Spanish, and more. Inspired by world music and soulful interpretations, their eclectic style creates a captivating, multicultural listening experience.

![Les Catherines live](images/les-catherines-live.jpg)

## Our Music

From intimate acoustic arrangements to vibrant multilingual interpretations, our music blends styles, cultures, and voices. Each performance is an invitation to travel across languages, emotions, and musical traditions.

## Live Music

We perform regularly in the Boulder area and host community music gatherings.

[See upcoming events →](events)

## A Musical Community: The CBart Studios Open Mic

In addition to performing across Boulder County, Les Catherines also host the CBArt Studios Open Mic in Superior—creating a warm, welcoming space where local musicians and audiences come together.
